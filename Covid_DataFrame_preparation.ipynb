```python
# usuniÄ™cie kolumn, ktÃ³re nic nie mÃ³wiÄ…
covid.drop(columns=['MEDICAL_UNIT','USMER'], inplace =True)
covid
```

```python
# zmiana nazw kolumn
covid.rename(columns=
             {'PATIENT_TYPE':'HOSPITALIZED',
              'DATE_DIED':'DEAD',
              'INMSUPR': 'IMMUNOSUPPRESSED',
              'CLASIFFICATION_FINAL': 'COVID_TEST_RESULT',
              'TOBACCO': "SMOKER",
              'CARDIOVASCULAR': 'CARDIOVASCULAR_DISEASE',
              'RENAL_CHRONIC': 'CHRONIC_KIDNEY_DISEASE',
              'ICU': 'ADMITTED_TO_ICU'}, inplace=True)
covid
```

```python
# zmiana kolejnoÅ›ci kolumn
covid = covid.loc[:,
                  ['COVID_TEST_RESULT', 
                   'HOSPITALIZED',
                   'INTUBED',
                   'ADMITTED_TO_ICU',
                   'DEAD',
                   'AGE',
                   'SEX',
                   'PREGNANT',
                   'SMOKER',
                   'PNEUMONIA',
                   'COPD',
                   'ASTHMA',
                   'IMMUNOSUPPRESSED',
                   'CARDIOVASCULAR_DISEASE',
                   'HIPERTENSION',
                   'CHRONIC_KIDNEY_DISEASE',
                   'DIABETES',
                   'OBESITY',
                   'OTHER_DISEASE']]
```

```python
covid
```

```python
# zamiana wartoÅ›ci w kolumnie "DEAD" na odpowiedÅº czy zmarÅ‚: 1, czyli "True" zamiast daty Å›mierci, oraz 0 czyli "False" zamiast 9999-99-99 
covid['DEAD'] = covid['DEAD'].replace({'9999-99-99': 0})
covid['DEAD'] = covid['DEAD'].apply(lambda x: 1 if x != 0 else 0)
covid
```

```python
covid['SEX'] = covid['SEX'].replace({1: 'female', 2: 'male'})
covid
```

```python
# sprawdzenie ile jest unikatowych wartoÅ›ci (i jakie to sÄ… wartoÅ›ci) dla kaÅ¼dej kolumny
for column_title in covid.columns:
    print(column_title, covid[column_title].unique(), covid[column_title].nunique())
```

```python
# zamiana 2 na 0 ('False"), 97, 98 i 99 na 'NaN' (Not a Number). WartoÅ›ci 'NaN' sÄ… automatycznie wykluczane przy tworzeniu wykresÃ³w.
covid[['INTUBED','ADMITTED_TO_ICU','PREGNANT','SMOKER','PNEUMONIA','COPD','ASTHMA','IMMUNOSUPPRESSED',
       'CARDIOVASCULAR_DISEASE','HIPERTENSION','CHRONIC_KIDNEY_DISEASE', 'DIABETES','OBESITY','OTHER_DISEASE']] = covid[[
           'INTUBED','ADMITTED_TO_ICU','PREGNANT','SMOKER','PNEUMONIA','COPD','ASTHMA','IMMUNOSUPPRESSED',
           'CARDIOVASCULAR_DISEASE','HIPERTENSION', 'CHRONIC_KIDNEY_DISEASE','DIABETES','OBESITY','OTHER_DISEASE']].replace({2: 0, 97: 'NaN', 98: 'NaN', 99: 'NaN'})
```

```python
# tu znaczenie 1 i 2 jest odwrotne niÅ¼ w reszcie kolumn , dlatego zamieniam te wartoÅ›ci osobno
covid['HOSPITALIZED'] = covid['HOSPITALIZED'].replace({1: 0, 2: 1})
covid
```

```python
covid['COVID_TEST_RESULT'] = covid['COVID_TEST_RESULT'].apply(lambda x: 1 if x in [1,2,3] else 0)
covid.sample(30)
```

```python
for column_title in covid.columns:
    print(column_title, covid[column_title].unique(), covid[column_title].nunique())
```
